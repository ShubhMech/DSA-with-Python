```python
class Graph():
    def __init__(self,dict={}):
        self.dict = dict
    
    def insert_edge(self, vertex, node):
        if vertex not in self.dict.keys():
            self.dict[vertex] = []
            
        self.dict[vertex].append(node)
        
    def Breadth_first_search(self,vertex):
        
        queue= [vertex]
        visited=[vertex]
            
        while len(queue)>0:
            
            dequeue = queue.pop(0)
            print(dequeue)
            for i in self.dict[dequeue]:
                if i not in visited:
                    visited.append(i)
                    queue.append(i)
                    
    def Depth_first_search(self,vertex):
        
        queue= [vertex]
        visited=[vertex]
            
        while len(queue)>0:
            
            dequeue = queue.pop()
            print(dequeue)
            for i in self.dict[dequeue]:
                if i not in visited:
                    visited.append(i)
                    queue.append(i)
                    
                    
        
        
g = Graph()
g.insert_edge('a','b')
g.insert_edge('b','c')
g.insert_edge('c','a')            
g.insert_edge('c','b')
g.insert_edge('c','d')
g.insert_edge('d','a')
```
